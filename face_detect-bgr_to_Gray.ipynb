```python
# detect if image is colored or gray scale
def detect_gray_colored(path):
    test = cv2.imread(path)
    im = Image.open(path).convert('RGB')
    w,h = im.size
    for i in range(w):
        for j in range(h):
            r,g,b = im.getpixel((i,j))
            if r != g != b: return False
    return True
```

```python
haar_face_cascade = cv2.CascadeClassifier("/opt/intel/computer_vision_sdk_2018.2.299/opencv/share/OpenCV/haarcascades/haarcascade_frontalface_alt.xml")
def read_face(path):
    colored_test = cv2.imread(path)
    test = cv2.cvtColor(colored_test, cv2.COLOR_BGR2GRAY)
    faces = haar_face_cascade.detectMultiScale(test, scaleFactor=1.3, minNeighbors=5)
    print ("Faces", len(faces))
    if (len(faces) == 1):
        for (x,y,w,h) in faces:
            cv2.rectangle(test, (x,y), (x+w, y+h), (0, 255, 0), 1)
            print (x, y, w, h)
            img = test[y:y+h-1, x:x+w-1]
            plt.imshow(img)
            return img
#     else:
#         print (path)
```
